```python
import pandas as pd
import joblib as jbl
```

```python
stats = pd.read_csv("../data/processed/rosstat.csv")
```

```python
region_mapping = jbl.load("../data/mappings/regions.jbl")
```

```python
sorted(set(region_mapping.keys()) - set(stats["region"]))
```

```python
sorted(set(stats["region"]) - set(region_mapping.keys()))
```

```python
rosstat_to_mts_mapping = {
    "Еврейская автономная область": "Еврейская АО",
    "Кемеровская область – Кузбасс": "Кемеровская область - Кузбасс",
    "г. Москва": "Москва",
    "Ненецкий автономный округ": "Ненецкий АО",
    "Республика Северная Осетия – Алания": "Республика Северная Осетия — Алания",
    "г. Санкт-Петербург": "Санкт-Петербург",
    "Ханты-Мансийский автономный округ – Югра": "Ханты-Мансийский АО — Югра",
    "Чукотский автономный округ": "Чукотский АО",
    "Ямало-Ненецкий автономный округ": "Ямало-Ненецкий АО",
}
```

```python
def map_region(reg: str) -> str:
    if reg in region_mapping:
        return reg
    elif reg in rosstat_to_mts_mapping:
        return rosstat_to_mts_mapping[reg]
    return None
```

```python
set(region_mapping.keys()) - set(stats["region"].map(map_region))
```

```python
stats
```

```python
stats.insert(loc=1, column="region_mts", value=stats["region"].map(map_region))
```

```python
stats.to_csv("../data/processed/rosstat-mts.csv", index=False)
```
